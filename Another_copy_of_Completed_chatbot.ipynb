```python
from sentence_transformers import SentenceTransformer
import faiss, numpy as np

embed_model = SentenceTransformer('all-MiniLM-L6-v2')
texts = df['text'].tolist()
embeddings = embed_model.encode(texts, convert_to_numpy=True, show_progress_bar=True)

dim = embeddings.shape[1]
index = faiss.IndexFlatL2(dim)
index.add(embeddings)

print('FAISS index ready with', index.ntotal, 'entries.')
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained('google/flan-t5-small')
model = AutoModelForSeq2SeqLM.from_pretrained('google/flan-t5-small')

def retrieve(query, k=3):
    q_emb = embed_model.encode([query], convert_to_numpy=True)
    D, I = index.search(q_emb, k)
    return [df.iloc[i]['text'] for i in I[0]]

def generate_answer(user_question, k=3):
    contexts = retrieve(user_question, k)
    context_text = '\n'.join([f'{i+1}) {c}' for i, c in enumerate(contexts)])
    prompt = f"""You are a friendly Sustainable Lifestyle Coach. Use the CONTEXT below to answer the USER QUESTION.
CONTEXT:
{context_text}

USER QUESTION:
{user_question}

Answer in a short, motivating coaching style with 2 practical steps and mention which context items you used."""

    inputs = tokenizer(prompt, return_tensors='pt', truncation=True)
    outputs = model.generate(**inputs, max_new_tokens=200)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Quick test
print(generate_answer('How can I reduce waste at home?'))
```

```python
import random
import pandas as pd
import datetime
import matplotlib.pyplot as plt

daily_challenges = [
    "Use a reusable water bottle instead of plastic ones.",
    "Unplug devices when not in use.",
    "Take a 10-minute walk instead of driving short distances.",
    "Switch off lights when leaving a room.",
    "Carry your own shopping bag.",
    "Eat a plant-based meal today.",
    "Avoid using the elevator – take the stairs!"
]

def get_daily_challenge():
    return random.choice(daily_challenges)

# ======================
# 🌍 Carbon Savings Tracker
# ======================
carbon_data = []

def log_action(user, action, carbon_saved):
    entry = {
        "Date": datetime.date.today(),
        "User": user,
        "Action": action,
        "Carbon_Saved (kg CO2)": carbon_saved
    }
    carbon_data.append(entry)

def show_carbon_summary():
    df = pd.DataFrame(carbon_data)
    if df.empty:
        print("No carbon savings logged yet.")
        return
    print(df)
    total = df["Carbon_Saved (kg CO2)"].sum()
    print(f"\n🌿 Total Carbon Saved: {total:.2f} kg CO2")
    df.groupby("Date")["Carbon_Saved (kg CO2)"].sum().plot(kind='bar')
    plt.title("Daily Carbon Savings")
    plt.xlabel("Date")
    plt.ylabel("CO2 Saved (kg)")
    plt.show()

# ======================
# 🏆 Gamification Leaderboard
# ======================
leaderboard = {}

def update_leaderboard(user, carbon_saved):
    leaderboard[user] = leaderboard.get(user, 0) + carbon_saved

def show_leaderboard():
    if not leaderboard:
        print("No leaderboard data yet.")
        return
    sorted_lb = dict(sorted(leaderboard.items(), key=lambda x: x[1], reverse=True))
    print("\n🏆 Sustainability Leaderboard:")
    for i, (user, score) in enumerate(sorted_lb.items(), start=1):
        print(f"{i}. {user} - {score:.2f} kg CO2 saved")

# ======================
# ⏰ Habit Reminders (Simulated)
# ======================
habit_reminders = [
    "Remember to turn off your tap while brushing!",
    "Did you carry your reusable bag today?",
    "Time to log your eco-friendly actions!",
    "Take a deep breath and step outside for a short walk!"
]

def send_habit_reminder():
    print(f"🔔 Reminder: {random.choice(habit_reminders)}")

# ======================
# 💬 Simple AI Chatbot (Text-based)
# ======================
def chatbot_response(user_input):
    user_input = user_input.lower()
    if "challenge" in user_input:
        return f"Today's eco-challenge 🌎: {get_daily_challenge()}"
    elif "log" in user_input:
        return "Type: log_action('YourName', 'Recycled plastic bottles', 0.5)"
    elif "leaderboard" in user_input:
        show_leaderboard()
        return ""
    elif "summary" in user_input:
        show_carbon_summary()
        return ""
    elif "reminder" in user_input:
        send_habit_reminder()
        return ""
    else:
        return "I'm your sustainability coach 🤖 — ask me for a 'challenge', 'summary', or 'reminder'!"

# ======================
# 🧩 Example Run
# ======================
print("🌍 Welcome to the AI Sustainability Coach!")
print("Type commands like: 'challenge', 'log', 'leaderboard', 'summary', or 'reminder'.")

# Example simulation:
print("\nExample Session ↓\n")
print(chatbot_response("challenge"))
log_action("Hariknee", "Used cloth bag", 0.3)
update_leaderboard("Hariknee", 0.3)
send_habit_reminder()
show_leaderboard()
```

```python
from sentence_transformers import SentenceTransformer
import faiss, numpy as np

embedder = SentenceTransformer("all-MiniLM-L6-v2")
texts = [f"{i['title']}. {i['text']}" for i in kb]
embs = embedder.encode(texts, convert_to_numpy=True)
index = faiss.IndexFlatL2(embs.shape[1])
index.add(embs)

def retrieve(query, k=3):
    q = embedder.encode([query], convert_to_numpy=True)
    D,I = index.search(q,k)
    return [kb[i] for i in I[0] if i!=-1]
```

```python
def format_fallback_reply(user_input, retrieved_texts):
    # Build a safe fallback answer using retrieved texts (no API needed)
    lines = []
    if retrieved_texts:
        # include top 1-2 retrieved tips
        for t in retrieved_texts[:2]:
            lines.append(f"- {t['title']}: {t['text']}")
        tip_block = "\n".join(lines)
        reply = f"Nice one! Here are a couple quick tips:\n{tip_block}\nMicro-challenge: try one of these for a week 🌿"
    else:
        reply = "Hey! 👋 I’m EcoBuddy — ask me for a tip (e.g., 'reduce food waste' or 'zero-waste skincare') and I’ll help with simple steps 🌱"
    return reply
```

```python
def generate_with_api(user_input, retrieved_texts, history):
    system_prompt = (
        "You are EcoBuddy — a Gen-Z sustainable lifestyle coach 🌿\n"
        "Personality: upbeat, friendly, and real. Use 1–2 emojis.\n"
        "- If greeted (hi, hello, hey, yo), greet back casually and ask how the user's eco vibe is.\n"
        "- If the topic isn’t about sustainability, connect it to eco-living in a chill way.\n"
        "- For sustainability questions, give 1 practical tip, 1 micro-challenge, and a positive sign-off.\n"
        "- Keep answers under ~80 words. Never invent facts; use retrieved info if relevant.\n"
    )
    if not retrieved_texts:
        retrieved_texts = [{"text":"General eco-coach responses."}]
    retrieved = "\n".join([f"{t['title']}: {t['text']}" for t in retrieved_texts])
    messages = [
        {"role":"system","content":system_prompt},
        {"role":"user","content": f"Context:\n{retrieved}\n\nUser said: {user_input}"}
    ]
    try:
        resp = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=messages,
            temperature=0.8,
            max_tokens=160
        )
        text = resp['choices'][0]['message']['content'].strip()
        return text
    except Exception as e:
        # Print traceback for debugging in Colab output, then fallback
        print("OpenAI API call failed — falling back to local/simple reply. Error:")
        traceback.print_exc()
        return format_fallback_reply(user_input, retrieved_texts)
```

```python
history = []

def chat_step(user_input):
    greetings = ["hi", "hello", "hey", "yo", "hola", "sup"]
    user_clean = user_input.lower().strip()
    if user_clean in greetings:
        reply = "Heyyy 👋 I’m EcoBuddy! How’s your eco vibe today? 🌿"
        history.append((user_input, reply))
        return reply
    r = retrieve(user_input, k=3)
    ans = generate_with_api(user_input, r, history)
    history.append((user_input, ans))
    return ans
```

```python
import gradio as gr

def gradio_chat(user_message, chat_state):
    reply = chat_step(user_message)
    chat_state = chat_state or []
    chat_state.append(("You", user_message))
    chat_state.append(("EcoBuddy 🌱", reply))
    return "", chat_state

with gr.Blocks() as demo:
    gr.Markdown("## 🌿 EcoBuddy — Your Gen-Z Sustainability Coach")
    chat = gr.Chatbot()
    msg = gr.Textbox(placeholder="Ask EcoBuddy anything...", label="Your message")
    state = gr.State([])
    send = gr.Button("Send")
    send.click(fn=gradio_chat, inputs=[msg, state], outputs=[msg, chat])
    gr.Examples(
        [["hi"],["How can I reduce food waste?"],["Give me an eco challenge"],["Tell me about plant care"]],
        inputs=msg
    )
demo.launch(share=True)
```
